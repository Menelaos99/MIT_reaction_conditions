```python
import numpy as np
import pandas as pd
from collections import defaultdict
import json
```

```python
filename = r'data/solid-state_dataset_20200713.json'

filedata = open(filename, mode='r').read()
jsonParse = json.loads(filedata)
```

```python
df=pd.read_json(filename)
df
```

```python
reactions=jsonParse['reactions']
reactions[1:5]
```

```python
from collections import defaultdict

targetDict = defaultdict(list)

identicals= defaultdict(int)
# Iterate over all reactions
for reaction in reactions:
    #using frozensets so that it can be hashed
    target = frozenset(reaction['targets_string'])

    precursors = reaction['precursors']
    #getting all precursors
    precursorSet = frozenset(precursor['material_formula'] for precursor in precursors)

    #getting duplicates out
    if(precursorSet in targetDict[target]):
        identicals[target]=identicals.get(target,0)+1
        continue

    targetDict[target].append(precursorSet)
```

```python
len(targetDict)
```

```python
multiplePrecursors=list()

for target, precursor in targetDict.items():
    if len(precursor)>1:
        multiplePrecursors.append(target)

len(multiplePrecursors)

# this number greatly increases when using list in the dictionary instead of set -> redundancy with same precursors and same target
```

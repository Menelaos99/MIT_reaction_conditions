```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

base_path = '/home/students/code/MLP/Menelaos/reaction_conditions/dataprocessing_MTE/dataset_v3/'


maes = []
for i in range(5):
    print('train' + str(i))
    df_train = pd.read_csv(base_path + 'train' + str(i) + '.csv')
    y_pred = np.mean(df_train['sintering_temp'])

    df_test = pd.read_csv(base_path + 'test' + str(i) + '.csv')
    y_true = df_test['sintering_temp']

    mae = np.mean(np.abs((y_true - [y_pred]*len(df_test))))
    maes.append(mae)
    plt.scatter(y_true, [y_pred]*len(df_test), label='Fold ' + str(i))
    plt.show()
```
